```python
# Count the number of missing values in each column
missing_values_count = data.isnull().sum()

# Filter columns with missing values
columns_with_missing_values = missing_values_count[missing_values_count > 0]

# Calculate the percentage of missing values in each column
total_entries = data.shape[0]
missing_values_percentage = (columns_with_missing_values / total_entries) * 100

# Create a new DataFrame to store the results
missing_values_df = pd.DataFrame({
    'Missing Values': columns_with_missing_values,
    '% of Missing Values': missing_values_percentage
})

# Sort the DataFrame by the number of missing values in descending order
missing_values_df = missing_values_df.sort_values(by='Missing Values', ascending=False)

# Display the DataFrame
missing_values_df
```

### Check for outliers 

```python
num_col = data.select_dtypes(include=['int64', 'float64']).columns

# Determine the number of columns in the DataFrame
num_columns = len(num_col)

# Calculate the number of rows required for subplots
num_rows = (num_columns + 2) // 3  # Add 2 to handle uneven number of columns

# Set up the subplot grid
fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, 5*num_rows))

# Flatten the axes array
axes = axes.flatten()

# Plot boxplots for each column
for i, column in enumerate(num_col):
    sns.boxplot(ax=axes[i], data=data, x = column, color='skyblue')
    axes[i].set_title(column)
    axes[i].set_xlabel('')

# Remove empty subplots if present
if num_columns < len(axes):
    for j in range(num_columns, len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

In the features 'Minumum Installs' and 'Maximum Installs' .there seem to be ectreme outlier . but cant say that they are logical outliers. 

### Deal with missing values.

```python
for i in data.columns :
    if i in num_col:
        data[i].fillna(value = data[i].median(), inplace = True)
    
    else :
        data[i].fillna(value = data[i].mode()[0] ,inplace = True)
        
data.isnull().sum()
```

5. Size 

```python
data['Size'].value_counts().sort_values(ascending=False)
```

```python
# Replace 'Varies with device' with NaN
data['Size'].replace({'Varies with device': np.nan},inplace =True)
data['Size'].isnull().sum()
```

```python
# Extract the suffixes from the 'Size' column
suffixes = data['Size'].str.lower().str.extract(r'([a-z]+)$', expand=False)

# Count the number of unique suffixes
unique_suffixes = suffixes.unique()

unique_suffixes
```

```python
data.Size.dtype
```

```python
data['Size']=data['Size'].astype(str)
# conver all the values in Megabytes
def convert_to_mb(value) :
    if (value.endswith('K') or value.endswith('k') ):
        value = float(value[:-1].replace(',',''))
        return value/1024
    if (value.endswith('G') or value.endswith('g') ):
        value = float(value[:-1].replace(',',''))
        return value*1024
    if (value.endswith('M') or value.endswith('m') ):
        value = float(value[:-1].replace(',',''))
        return value
    if (value == 'nan' ):
        value = np.nan
        return value
    else  : 
        return value.replace(',','')
    
# Apply the function to create the 'Size in MB' column
data['Size in MB'] = data['Size'].apply(convert_to_mb)
```

```python
data['Size in MB'].isnull().sum()
```

```python
data['Size in MB'].dtype
```

```python
data['Size in MB'].fillna(value = data['Size in MB'].median() ,inplace=True)
data['Size in MB'].isnull().sum()
```
